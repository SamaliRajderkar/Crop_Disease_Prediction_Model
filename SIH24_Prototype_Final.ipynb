```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import cv2
```

```python
data_dir=r"C:\Users\Admin\OneDrive\Documents\Crop Disease Prediction SIH24\SIH_FINAL\plantvillage dataset"
```

```python
data = tf.keras.utils.image_dataset_from_directory(data_dir, batch_size=32, label_mode='categorical')
```

```python
classes=data.class_names
classes
```

```python
# Normalize the dataset
data = data.map(lambda x, y: (x / 255.0, y))
```

```python
train_size = 500
val_size = 63
test_size = 150
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size + val_size).take(test_size)
```

```python
# Define the CNN model
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(16, (3, 3), activation='relu', input_shape=(256, 256, 3)),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(15, activation='softmax')
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python

# Train the model with validation
history = model.fit(train, validation_data=val, epochs=10)
```

```python
# Save the model
model.save("custom_trained_model.h5")
```

```python
# Evaluate on the test set
test_loss, test_accuracy = model.evaluate(test)
print(f"Test accuracy: {test_accuracy}")
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Load your dataset
data_dir = r"C:\Users\Admin\OneDrive\Documents\Crop Disease Prediction SIH24\SIH_FINAL\plantvillage dataset"
data = tf.keras.utils.image_dataset_from_directory(data_dir, batch_size=32, label_mode='categorical')

# Normalize the dataset
data = data.map(lambda x, y: (x / 255.0, y))

# Define sizes
train_size = 500
val_size = 63
test_size = 150

# Split the dataset
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size + val_size).take(test_size)

# Load the model
model = load_model("custom_trained_model.h5")

# Prepare the test data
x_test = []
y_test = []
for batch in test.as_numpy_iterator():
    x, y = batch
    y = y.astype(int)
    x_test.append(x)
    y_test.append(y)

x_test = np.concatenate(x_test)
y_test = np.concatenate(y_test)

# Make predictions on the test set
yhat = model.predict(x_test)

# Get the predicted labels
y = np.argmax(yhat, axis=1)

# Now plot the histogram of predicted labels
plt.figure(figsize=(10, 7))
plt.hist(y, bins=np.arange(len(data.class_names) + 1) - 0.5, edgecolor='black', alpha=0.7)
plt.xticks(np.arange(len(data.class_names)), data.class_names, rotation=45)
plt.xlabel('Predicted Labels')
plt.ylabel('Frequency')
plt.title('Histogram of Predicted Labels')
plt.grid(axis='y')
plt.show()
```
